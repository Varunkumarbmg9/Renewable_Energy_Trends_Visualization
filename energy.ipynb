# Overview
This dataset provides a comprehensive overview of renewable energy production and consumption across various sectors. The data is organized by year, month, and sector, with specific attributes detailing energy contributions from different renewable sources.

# Attributes
- Year: The year when the data was recorded (e.g., 1973).
- Month: The month when the data was recorded, represented as an integer (1 = January, 2 = February, etc.).
- Sector: The sector consuming or producing the energy (e.g., Commercial, Residential, Industrial, Electric Power, Transportation).
- Hydroelectric Power: Energy generated through hydroelectric means.
- Geothermal Energy: Energy produced from geothermal sources.
- Solar Energy: Energy harnessed from solar panels or similar technologies.
- Wind Energy: Energy generated from wind turbines.
- Wood Energy: Energy derived from wood and wood-based biomass products.
- Waste Energy: Energy produced from the combustion or processing of waste materials.
- Fuel Ethanol (Excluding Denaturant): Ethanol used as a renewable fuel, excluding denaturants.
- Biomass Losses and Co-products: By-products or energy losses during biomass energy production.
- Biomass Energy: Aggregated energy produced from biomass sources.
- Total Renewable Energy: Sum of all renewable energy sources in the dataset.
- Renewable Diesel Fuel: Diesel fuel derived from renewable materials.
- Other Biofuels: Biofuels that do not fit into specific categories above.
- Conventional Hydroelectric Power: Energy produced by conventional hydroelectric methods.
- Biodiesel: A renewable substitute for traditional diesel fuel.

```python
# Check for missing values in each column
missing_values = data.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```python
# Rename columns for easier processing (remove spaces, standardize names)
data.columns = [col.strip().replace(" ", "_").lower() for col in data.columns]
```

```python
# Check for duplicate rows
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
# Dataframe summary

def summary(data):
    print(f'data shape: {data.shape}')
    summ = pd.DataFrame(data.dtypes, columns=['Data Type'])
    summ['Missing#'] = data.isna().sum()
    summ['Missing%'] = (data.isna().sum())/len(data)
    summ['Dups'] = data.duplicated().sum()
    summ['Uniques'] = data.nunique().values
    summ['Count'] = data.count().values
    desc = pd.DataFrame(data.describe(include='all').transpose())
    summ['Min'] = desc['min'].values
    summ['Max'] = desc['max'].values
    summ['Average'] = desc['mean'].values
    summ['Standard Deviation'] = desc['std'].values
    summ['First Value'] = data.loc[0].values
    summ['Second Value'] = data.loc[1].values
    summ['Third Value'] = data.loc[2].values

    display(summ)

summary(data)
```

# No Missing, No duplicates.

```python
# The 2024 data is incomplete. So, let's delete them.

index_names = data[ data['year'] == 2024 ].index

data.drop(index_names, inplace = True)
```

# EDA

```python
# Columns, Sectors and Colors

cols = ['hydroelectric_power', 'geothermal_energy', 'solar_energy', 'wind_energy',
       'waste_energy', 'fuel_ethanol,_excluding_denaturant', 'biomass_losses_and_co-products',
       'renewable_diesel_fuel', 'other_biofuels', 'conventional_hydroelectric_power',
       'biodiesel']

sectors = ['Commerical', 'Electric Power', 'Industrial', 'Residential',
           'Transportation']

colors = ["#8c0404","#f25ed0","#21618C","#16A085","#34495E"]
```

```python
# Energy per Year

fig, axes = plt.subplots(11, 1, figsize=(8, 40))

for ind, col in enumerate(cols):
    plt.subplot(11, 1, ind+1)
    df_top = df.groupby(['year'])[[col]].sum().reset_index()
    df_top = df_top.sort_values(by=(col), ascending=False)

    sns.barplot(x=df_top["year"], y=df_top[col], palette="Blues")
    plt.yticks(rotation=45, fontsize = 7)
    plt.xticks(rotation=90, fontsize = 7)   
    plt.title(f"Consumption per Year according to {col}", fontsize = 12, color = 'blue')

plt.tight_layout()
```

```python
# Year-on-Year Consumption of Different Types of Energy

df_year = df.groupby('year').sum().reset_index().drop(columns=['month', 'sector'])
plt.figure(figsize=(12, 6))
for energy_type in cols:
    plt.plot(df_year['year'], df_year[energy_type], label=energy_type)

plt.xlabel('year')
plt.ylabel('Energy Consumption')
plt.title(f'Year-on-Year Consumption of Different Types of Energy',  fontsize = 15, color = 'blue')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

```python
 # Wind / Solar vs Conventional Hydroelectric Power
    
df_year = df.groupby('year').sum().reset_index().drop(columns=['month', 'sector'])

# Plot
df_year.plot(x="year", y=['wind_energy', 'conventional_hydroelectric_power', 'solar_energy'],
        kind="line", figsize=(10, 6))
plt.title("Wind / Solar vs Conventional Hydroelectric Power", fontsize = 16, color = 'blue')
plt.ylabel('Consumption', rotation=90, fontsize = 10)
plt.annotate('Surpasses',
      horizontalalignment = 'left',
      verticalalignment = 'bottom',
      xytext = (2012, 1200),
      xy = (2018.5, 1000),
      arrowprops = dict(facecolor = 'red', shrink = 0.05))
plt.annotate('Surpasses',
      horizontalalignment = 'left',
      verticalalignment = 'bottom',
      xytext = (2020, 1100),
      xy = (2022.5, 850),
      arrowprops = dict(facecolor = 'yellow', shrink = 0.05))

plt.grid()
plt.show()
```

```python
# Energy Consumption by Sectors over the years

for i in sectors:
    
    if i == 'Commerical':
        df_sec = df[df['sector'] == 'Commerical']
        df_sec = df_sec.groupby('year')[cols].sum().reset_index()

        df_sec.plot(x="year", y=cols, kind="line", figsize=(8, 5))
        plt.title("Energy Consumption by Commerical over the years", fontsize = 14, color = 'blue')
        plt.ylabel('Energy Consumption', rotation=90, fontsize = 10)
        plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

        plt.grid()
        plt.show()
    else:
        df_sec = df[df['sector'] == i]
        df_sec = df_sec.groupby('year')[cols].sum().reset_index()

        df_sec.plot(x="year", y=cols, kind="line", figsize=(8, 5))
        plt.title(f"Energy Consumption by {i} over the years", fontsize = 14, color = 'blue')
        plt.ylabel('Energy Consumption', rotation=90, fontsize = 10)
        plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

        plt.grid()
        plt.show()
```

```python
# Energy Sources by Sector in 2023

df_2023 = df[df['year'] == 2023]

plt.figure(figsize=(12, 6))
bottom = np.zeros(len(df_2023))
for source in cols:
    plt.bar(df_2023['sector'], df_2023[source], bottom=bottom, label=source)
    bottom += df_2023[source]
plt.xlabel('Sector')
plt.ylabel('Energy')
plt.title(f'Energy Sources by Sector in 2023')
plt.legend(bbox_to_anchor=(1.02, 1))
plt.tight_layout()
plt.show()
```
